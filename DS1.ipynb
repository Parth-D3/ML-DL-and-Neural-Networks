#### The dataset has '?' as a placeholder for null values. We will replace it with np.NaN

```python
df.replace(to_replace = '?', value = np.NaN, inplace = True)
```

#### Checking dataset after replacement

```python
df.head(10)
```

#### Calculating null values after replacement

```python
df.isnull().sum()
```

#### Replaced null values with mean for column 'NormalizedLosses'

```python
df['NormalizedLosses'] = df['NormalizedLosses'].astype('float')
```

```python
df['NormalizedLosses'].fillna(122,inplace = True)
```

```python
df.isnull().sum()
```

#### Replaced null values for columns 'bore','stroke','HorsePower' and 'PeakRPM' with respectvie mean

```python
temp_list = ['Bore', 'Stroke', 'HorsePower', 'PeakRPM']
for colName in temp_list:
    # converting columns to float dtype
    df[colName] = df[colName].astype('float')
    
    # filling nan values with mean
    df[colName].fillna(df[colName].mean(), inplace = True)

df.describe()
```

```python
df.isnull().sum()
```

#### Replaced null values in column 'NumDoors' with highest freqency entry i.e four

```python
df['NumDoors'].fillna('four',inplace = True)
df.isnull().sum()
```

#### Dropping null entries in Price column as Price is target variable and it shouldnt have null values

```python
df.dropna(subset = ['Price'],inplace = True)
df['Price'] = df['Price'].astype('float')
```

#### Adding new columns for city-L/100km and highway=L/100km

```python
def convert(colName):
    for i in df[colName]:
        df.loc[df[colName]==i,colName + '_L/100Km'] = 235 / i
```

```python
convert('CityMPG')
convert('HighWayMPG')
```

```python
df.head(10)
```

```python
df.rename(columns = {'CityMPG_L/100Km':'City_L/100km', 'HighWayMPG_L/100Km':'Highway_L/100Km'}, inplace = True)
```

#### Normalizing the columns 'Length', 'Width', 'Height'

```python
for col in ['Length', 'Width', 'Height']:
    m1 = df[col].max()
    df[col] = df[col]/m1
    
df.head()
```

#### Encoding -> Converting categorical data to numeric

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
df[df.select_dtypes(include = ['object']).columns] = df[df.select_dtypes(include = ['object']).columns].apply(le.fit_transform)
```
